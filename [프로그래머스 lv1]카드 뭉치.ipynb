```python
# cards1 = ["i", "drink", "water"]
cards1 = ["i", "water", "drink"]
cards2 = 	["want", "to"]	
goal = 	["i", "want", "to", "drink", "water"]
```

```python
def solution(cards1, cards2, goal):
    answer = 'Yes'
    x, y = 0, 0
    for word in goal:
        if x < len(cards1) and cards1[x] == word:
            x+=1
        elif y < len(cards2) and cards2[y] == word:
            y+=1
        else:
            return "No"
            
    return answer

```

```python
def solution1(cards1, cards2, goal):
    answer = 'Yes'
    for word in goal:
        if len(cards1)>0 and cards1[0] == word:
            del cards1[0]
        elif len(cards2)>0 and cards2[0] == word:
            del cards2[0]
        else:
            return "No"
            
    return answer
```
