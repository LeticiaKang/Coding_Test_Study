```python
n, m, section, result = 8,	4,	[2, 3, 6],	2
# n, m, section, result = 5,	4,	[1, 3],	1
# n, m, section, result = 4,	1,	[1, 2, 3, 4],	4
```

```python
def solution(n, m, section):
    answer, temp = 0, 0
    for s in section:
        if s > temp:
            answer+=1
            temp = s + (m-1)
    return answer
```

```python
print(solution(8,	4,	[2, 3, 6]))
print(solution(5,	4,	[1, 3]))
print(solution(4,	1,	[1, 2, 3, 4]))
```
