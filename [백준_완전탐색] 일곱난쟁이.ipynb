**입력**   
아홉 개의 줄에 걸쳐 난쟁이들의 키가 주어진다.   
주어지는 키는 100을 넘지 않는 자연수이며, 아홉 난쟁이의 키는 모두 다르며, 가능한 정답이 여러 가지인 경우에는 아무거나 출력한다.

```python
data = []
temp1, temp2 = 0, 0

for _ in range(9):
    data.append(int(input()))

sum_data = sum(data)

for x in range(8):
    for y in range(x+1,9):
        if (sum_data - data[x] - data[y]) == 100:
            temp1 = data[x]
            temp2 = data[y]
            
data.remove(temp1)
data.remove(temp2)
data.sort()
            
for z in data:
    print(z)
```
