```python
def solution(keymap, targets):
    answer = []
    # 목표값 하나씩 빼오기
    for target in targets:
        count = 0
        # 목표값에서 글자 하나씩 가져오기
        for t in target:
            bool = False
            temp = 101
            # keymap에서 1)글자가 있는지, 
            # 2)몇 번 눌러야 글자가 나오는지, 
            # 3)둘다 가지고 있으면 작은 값은 어떤 거인지 확인행 함
            for key in keymap:
                if t in key:
                    temp = min(key.index(t), temp)
                    bool = True
            if not bool:           
                temp = -1
                break   
            
            count += temp          
            
        answer.append(count)  
            
    return answer
```

```python
def solution(keymap, targets):
    answer = []
     
    for target in targets:
        count = 0
        
        for t in target:
            con = False
            temp = 101
            
            for key in keymap:
                if t in key:
                    temp = min(key.index(t)+1, temp)
                    con = True
                    
            if not con:           
                count = -1
                break
                
            count += temp   
            
        answer.append(count)  
            
    return answer
```

```python
print(solution(["ABACD", "BCEFD"]	, ["ABCD","AABB"]))
print(solution(["AA"]	, ["B"]))
print(solution(["AGZ", "BSSS"]	, ["ASA","BGZ"]))
```
