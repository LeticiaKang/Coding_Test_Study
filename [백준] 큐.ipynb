문제 링크: https://www.acmicpc.net/problem/10845

**입력**   
첫째 줄에 주어지는 명령의 수 N (1 ≤ N ≤ 10,000)이 주어진다. 둘째 줄부터 N개의 줄에는 명령이 하나씩 주어진다. 주어지는 정수는 1보다 크거나 같고, 100,000보다 작거나 같다. 문제에 나와있지 않은 명령이 주어지는 경우는 없다.

```python
import sys

n = int(sys.stdin.readline().strip())

data = []

for _ in range(n):
    temp = sys.stdin.readline().strip().split()
    
    if temp[0] == 'push':
        data.append(temp[1])
        
    elif temp[0] == 'pop':
        print(data.pop(0)) if len(data) != 0 else print(-1)
        
    elif temp[0] == 'size':
        print(len(data))
        
    elif temp[0] == 'empty':
        print(0) if len(data) != 0 else print(1)
        
    elif temp[0] == 'front':
        print(data[0]) if len(data) != 0 else print(-1)
        
    elif temp[0] == 'back':
        print(data[-1]) if len(data) != 0 else print(-1)
```
