### 문제 이름
바탕화면 정리

### 링크

https://school.programmers.co.kr/learn/courses/30/lessons/161990

### 풀이

```python
def solution(wallpaper):
    answer = []
    x1, x2 = 0, 0
    y1, y2 = 0, 1
    
    for num, wall in enumerate(wallpaper):
        if '#' in wall:
            y1 = wall.index('#')
            x1 = num
            break
    
    for num, wall in enumerate(wallpaper[::-1]):
        wall_reverse = wall[::-1]
        if '#' in wall_reverse:
            x2 = len(wallpaper) - num  
            break
            
    for wall in wallpaper:
        temp1 =  wall.find('#')
        temp2 = len(wall) - wall[::-1].find('#')
        
        if temp1==-1:
            temp1 = y1
            temp2 = y2
        
        if temp1 < y1: 
            y1 = temp1
            
        if temp2 > y2:
            y2 = temp2 
        
        
    return [x1, y1, x2, y2]
```

### 다른 사람 풀이
- 인덱싱 이용하기

```python
def solution(wallpaper):
    a, b = [], []
    for i in range(len(wallpaper)):
        for j in range(len(wallpaper[i])):
            if wallpaper[i][j] == "#":
                a.append(i)
                b.append(j)
    return [min(a), min(b), max(a) + 1, max(b) + 1]
```
