### (두번째 풀이)

```python
def solution(players, callings):
    # players : 1등부터 등수 순서대로 이름이 담김(중복 없음, 소문자만, 3자 이상 10자 이하)
    # callings : 해설진이 부른 이름이 있음(2개 이상 백만 이하, 1등인 선수 이름은 불리지 않음)
    # answer : 경주가 끝났을 때 1등부터의 선수들 이름
    answer = []
    result = {player: i for i, player in enumerate(players)} # 선수: 등수
    
    for name in callings:
        idx = result[name] # 원래 위치 저장
        result[name] -= 1 # 호명되었기 때문에 앞으로 한칸 와야 한다.
        result[players[idx-1]] += 1 # 호명된 이름 앞에 있던 사람은 1칸 뒤로 와야 함
        players[idx-1], players[idx] = players[idx], players[idx-1] # 위치 변경
    
        
    return players
```

```python
players, callings = ["mumu", "soe", "poe", "kai", "mine"], ["kai", "kai", "mine", "mine"]

solution(players, callings)
```
