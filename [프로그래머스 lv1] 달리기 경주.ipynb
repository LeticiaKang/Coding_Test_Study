```python
def solution(players, callings):
    # players : 1    ( , , 3  10 )
    # callings :    (2   , 1    )
    # answer :    1  
    answer = []
    result = {player: i for i, player in enumerate(players)} # : 
    
    for name in callings:
        idx = result[name] #   
        result[name] -= 1 #      .
        result[players[idx-1]] += 1 #      1   
        players[idx-1], players[idx] = players[idx], players[idx-1] #  
    
        
    return players
```
